```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
def normalize_and_standardize_columns(df, cols_to_transform):
    for col in cols_to_transform:
        # Standardize the column
        scaler_standard = StandardScaler()
        df[f'{col}'] = scaler_standard.fit_transform(df[[col]])

        # Min-Max Normalize the column
        scaler_minmax = MinMaxScaler()
        df[f'{col}'] = scaler_minmax.fit_transform(df[[col]])
    return df
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from scipy.stats import spearmanr


# Define features (X) and target (y)
features = [
    'Screen Assists',    # Rudy Gobert is a good example for making lots of screen assists => meaning: adding position would better explain the whole model
    'Loose Balls Recovered',
    'Contested Shots',
    'Age',

]
target = 'min_per_game'

X = combined_hustlers_1[features]
y = combined_hustlers_1[target]

# Split the data into training and testing sets with a 60/40 split
X_train_gb, X_test_gb, y_train_gb, y_test_gb = train_test_split(X, y, test_size=0.40, random_state=42)

# Initialize and train the Gradient Boosting Regressor model
gbr_model1 = GradientBoostingRegressor(random_state=42) # Using default parameters, can be tuned
gbr_model1.fit(X_train_gb, y_train_gb)

# Make predictions on the test set
y_pred_gb = gbr_model1.predict(X_test_gb)

# Evaluate the model
mse_gb_1 = mean_squared_error(y_test_gb, y_pred_gb)
r2_gb_1 = r2_score(y_test_gb, y_pred_gb)
spearman_gb_1, _ = spearmanr(y_test_gb, y_pred_gb)

print(f"Gradient Boosted Regression Tree Model Performance:")
print(f"Mean Squared Error (MSE): {mse_gb_1:.4f}")
print(f"R-squared (R2): {r2_gb_1:.4f}")
print(f"Spearman Correlation: {spearman_gb_1:.4f}")

# Optionally, display feature importances
print("\nFeature Importances:")
for feature, importance in zip(features, gbr_model1.feature_importances_):
    print(f"{feature}: {importance:.4f}")
```

```python
def creating_quantile_indices_loop(df):
    columns = ['avg_talent_index', 'avg_hustler_index']

    for col in columns:
        first_third = df[col].quantile(0.33)
        second_third = df[col].quantile(0.66)
        df[f'{col}_third'] = ''  # initialize new column

        for i in range(len(df)):
            if df.loc[i, col] < first_third:
                df.loc[i, f'{col}_third'] = 'low'
            elif df.loc[i, col] < second_third:
                df.loc[i, f'{col}_third'] = 'medium'
            else:
                df.loc[i, f'{col}_third'] = 'high'

    return df
```

```python
from sklearn.metrics import brier_score_loss
import numpy as np
import pandas as pd
from pgmpy.inference import VariableElimination
from pgmpy.estimators import BayesianEstimator
from pgmpy.models import DiscreteBayesianNetwork

brier_results = []

test_years = sorted(final_merged_df_2["Year"].unique())

# Define categorical features and target once
categorical_features = ['n_superstars', 'n_stars', 'avg_talent_index_third', 'avg_hustler_index_third']
target_variable = 'Play_off_deepness'

# Get all possible categories from the full dataset
all_n_superstars_categories = final_merged_df_2['n_superstars'].unique().tolist()
all_n_stars_categories = final_merged_df_2['n_stars'].unique().tolist()
all_avg_talent_index_third_categories = ['low', 'medium', 'high']
all_avg_hustler_index_third_categories = ['low', 'medium', 'high']
all_play_off_deepness_categories = [0, 1, 2]

for year in test_years:

    train_df = final_merged_df_2[final_merged_df_2["Year"] < year].copy()
    test_df  = final_merged_df_2[final_merged_df_2["Year"] == year].copy()

    if train_df.empty or test_df.empty:
        print(f"Skipping {year} (missing data)")
        continue

    print(f"Evaluating year {year}...")

    # Ensure categorical typing
    for df in [train_df, test_df]:
        df['n_superstars'] = pd.Categorical(df['n_superstars'], categories=all_n_superstars_categories)
        df['n_stars'] = pd.Categorical(df['n_stars'], categories=all_n_stars_categories)
        df['avg_talent_index_third'] = pd.Categorical(df['avg_talent_index_third'], categories=all_avg_talent_index_third_categories)
        df['avg_hustler_index_third'] = pd.Categorical(df['avg_hustler_index_third'], categories=all_avg_hustler_index_third_categories)
        df['Play_off_deepness'] = pd.Categorical(df['Play_off_deepness'], categories=all_play_off_deepness_categories)

    # Reinitialize and fit model for this year
    model_year_instance = DiscreteBayesianNetwork([
        ('n_superstars', 'Play_off_deepness'),
        ('n_stars', 'Play_off_deepness'),
        ('avg_talent_index_third', 'Play_off_deepness'),
        ('avg_hustler_index_third', 'Play_off_deepness')
    ])

    model_year_instance.fit(data=train_df, estimator=BayesianEstimator)
    infer = VariableElimination(model_year_instance)

    true_labels = []
    all_probs = []

    # ---- Prediction loop ----
    for _, row in test_df.iterrows():

        evidence = {
            "n_superstars": row["n_superstars"],
            "n_stars": row["n_stars"],
            "avg_talent_index_third": row["avg_talent_index_third"],
            "avg_hustler_index_third": row["avg_hustler_index_third"]
        }

        q = infer.query(variables=["Play_off_deepness"], evidence=evidence)
        probs = q.values       # array of three values
        all_probs.append(probs)

        # ---- FIX: Convert NaN truth labels to NO_PLAYOFFS=0 ----
        if pd.isna(row["Play_off_deepness"]):
            clean_label = 0
        else:
            clean_label = int(row["Play_off_deepness"])

        true_labels.append(clean_label)

    # Convert to arrays
    all_probs = np.array(all_probs)
    true_labels = np.array(true_labels)

    # Compute Brier scores per class
    class_briers = []
    for cls in range(3):
        y_bin = (true_labels == cls).astype(int)
        brier = brier_score_loss(y_bin, all_probs[:, cls])
        class_briers.append(brier)

    mean_brier = float(np.mean(class_briers))
    median_brier = float(np.median(class_briers))

    # Save results
    brier_results.append({
        "year": year,
        "mean_brier": mean_brier,
        "median_brier": median_brier,
        "class_brier": class_briers,
        "predicted_probs": all_probs,
        "true_labels": true_labels
    })

# Convert list of dicts to DataFrame
brier_results = pd.DataFrame(brier_results)
print("Finished evaluation")
brier_results
```
